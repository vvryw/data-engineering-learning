## Write pandas dataframe to JSON file

```python
import glob
from pathlib import Path

RESOURCE_DIR = Path("../../../resources/data")
```

```python
src_file_names = glob.glob(
    str(RESOURCE_DIR / "retail_db" / "*" / "part-*" ),
    recursive=True
)
src_file_names
```

### Get schemas

```python
import json

schemas = json.load(open(RESOURCE_DIR / "retail_db/schemas.json"))
```

### Get columns function

```python
def get_column_names(schemas, ds_name, sorting_key='column_position'):
    columns_details = schemas[ds_name]
    columns = sorted(columns_details, key=lambda col: col[sorting_key])
    return [col['column_name'] for col in columns]
```

### Get shape of file

```python
import re
import pandas as pd

for file in src_file_names:
    print(f'processing: {file}')
    file_path_list = re.split('[/\\\]', file)
    ds_name = file_path_list[-2]

    if ds_name == 'orders_json':
        print('skip orders_json files')
        continue

    columns = get_column_names(schemas, ds_name)
    df = pd.read_csv(file, names=columns)
    
    print(f'shape of {ds_name} is {df.shape}')
```

### Generate file paths for json files

```python
tgt_base_dir = 'retail_db_json'
```

```python
for file in src_file_names:
    file_path_list = re.split('[/\\\]', file)
    ds_name = file_path_list[-2]
    
    if ds_name == 'orders_json':
        print('skip orders_json files')
        continue

    file_name = file_path_list[-1]
    json_file_path = f'../../../resources/{tgt_base_dir}/{ds_name}/{file_name}'
    print(json_file_path)
```

### Save JSON files

```python
import os

for file in src_file_names:
    print(f'processing: {file}')
    file_path_list = re.split('[/\\\]', file)
    ds_name = file_path_list[-2]

    if ds_name == 'orders_json':
        print('skip orders_json files')
        continue

    file_name = file_path_list[-1]
    json_file_path = f'../../../resources/data/{tgt_base_dir}/{ds_name}/{file_name}'
    columns = get_column_names(schemas, ds_name)
    df = pd.read_csv(file, names=columns)
    
    os.makedirs(f'../../../resources/data/{tgt_base_dir}/{ds_name}' , exist_ok=True)

    df.to_json(
        json_file_path,
        orient='records',
        lines=True
    )

    print(f'shape of {ds_name} is {df.shape}')
    print(f'save to: {json_file_path}')
```
