### Lambda function
- lambda = function with no name

```python
sumNl = lambda n: (n * (n + 1)) / 2
```

```python
sumNl(5)
```

```python
[(n * (n + 1)) / 2 for n in l]
```

```python
def sumN(n):
    return (n * (n + 1)) / 2

list(map(sumN, l))
```

```python
list(map(lambda n: (n * (n + 1)) / 2, l))
```
