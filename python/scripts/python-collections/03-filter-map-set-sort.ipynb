### Filter

```python
filter(lambda order: order.split(',')[3] == "COMPLETE" , orders)
```

```python
list(filter(lambda order: order.split(',')[3] == "COMPLETE" , orders))

# filter only True values
# for every elements in orders
# element in orders -> lambda function
# lambda function -> split [3] == COMPLETE
```

```python
list(filter(lambda order: order.split(',')[3] in ("COMPLETE", "CLOSED") , orders))
```

## Get unique values from list using Map and Set

### Map

```python
map?
```

```python
list(map(lambda order: order.split(',')[3], orders))

# len = 20
```

### Set
* get unique values

```python
set(map(lambda order: order.split(',')[3], orders))
```

## Sort data by using Key

```python
sorted?
```

```python
order = orders[0]
```

```python
int(order.split(',')[2])
```

```python
sorted(orders)
# sortede by first key
```

```python
# sorted by 3rd element of orders (ASC)

sorted(orders, key=lambda order: int(order.split(',')[2]))
```

```python
# sorted by 3rd element of orders (DESC)

sorted(orders, key=lambda order: int(order.split(',')[2]), reverse=True)
```
