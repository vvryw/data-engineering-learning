### function

```python
def sumN(n):
    return (n * (n + 1)) / 2
```

```python
sumN(a)
```

```python
def sumRange(lower_bound, upper_bound):
    return sumN(upper_bound) - sumN(lower_bound - 1)
```

```python
sumRange(5, 10)
```
