## Python List

```python
sales = [125, 205, 183, 312]
# index  0    1    2    3
```

```python
type(sales)
```

### get from position

```python
print(sales[2])
print(sales[-1])    # last member and so on..
```

```python
print(sales[1:3])   # element 1 -> 3
print(sales[1:])    # element 1 -> last
print(sales[:3])    # element first -> 3
```

### string list

```python
cities = ['Dallas', 'Frisco', 'Hyderabad', 'Bengaluru']
```

```python
print(cities[2:])
print(cities[-2:])
```

### number of elements in list

```python
len(cities)
```
