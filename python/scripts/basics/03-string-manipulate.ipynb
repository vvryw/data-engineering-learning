## String Manipulate Functions

```python
order = '1,2022-01-01 00:00:00.0,1,COMPLETE'
```

```python
type(order)
```

## Split functions

```python
# order.split?
# Signature: order.split(sep=None, maxsplit=-1)

order.split()

# list >> ['1,2022-01-01', '00:00:00.0,1,COMPLETE']
```

### split by character

```python
order.split(sep=',')

# list >> ['1', '2022-01-01 00:00:00.0', '1', 'COMPLETE']
```

### get split output by index

```python
print(order.split(sep=',')[0])     # index = 0: '1'
print(order.split(sep=',')[1])     # index = 1
```

```python
order.split(sep=',')[3].lower()
```
