## Read CSV data using Pandas

```python
import pandas as pd
```

```python
orders_columns = [
    'order_id', 'order_date', 'order_customer_id', 'order_status'
]
```

```python
orders_data_path = '../../../resources/data/retail_db/orders/part-00000'

pd.read_csv(orders_data_path).head()
```

```python
pd.read_csv(orders_data_path, header=None).head()
```

Add columns name

```python
pd.read_csv(
    orders_data_path,
    names=orders_columns
).head()
```

## Query example

```python
orders = pd.read_csv(orders_data_path, names=orders_columns)

orders.head()
```

```python
orders.columns
```

```python
orders['order_status'].unique()
```

```python
orders.query('order_status == "COMPLETE"')
```

```python
orders.query('order_status == "COMPLETE" and order_date == "2014-01-01 00:00:00.0" ').head()
```

```python
orders.query('order_status == "COMPLETE" or order_status == "CLOSED" ')
```

```python
orders.query('order_status == ("COMPLETE" , "CLOSED") ')
```
