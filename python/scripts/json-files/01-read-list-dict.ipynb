## Read JOSN string to Python dicts or lists

### JSON document

```python
order_jd = {
    "order_id": 1,
    "order_date": 
    "2013-07-25"
}
```

```python
order_jd
```

```python
order_jd = "{\"order_id\": 1, \"order_date\": \"2013-07-25\"}"
```

```python
order_jd
```

JSON document -> convert -> dict

```python
import json
```

```python
order = json.loads(order_jd)
```

```python
type(order)
```

```python
order['order_date']
```

If main string use single quote -> no need to escape

```python
order_jd = '{"order_id": 1, "order_date": "2013-07-25"}'
```

```python
json.loads(order_jd)['order_date']
```

### JSON Array

```python
cities_ja = '[\"Dallas\", \"Frisco\", \"Hyderabad\"]'
# optional to escape
```

```python
cities_ja
```

JSON array -> convert -> list

```python
cities = json.loads(cities_ja)
```

```python
type(cities)
```

```python
cities[1]
```
