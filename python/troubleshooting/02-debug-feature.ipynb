## Debugging vscode notebook using Debug feature

```python
def sumInteger(l):
    total = 0
    for i in l:
        if type(i) == 'int':  # type int should not be 'int' => string
            total += i

    return total
```

```python
sumInteger([1, 2, 'Hello', None, 3.5, 4, 3])
```
